***Imports & Declaration***

```python
import matplotlib.pyplot as plt
import numpy as np



primeArr = []  # example array
primes = []
result = [0]  # to store the differences
```

**generate_primes**

```python
def generate_primes(n):
    
    for num in range(2, n+1):
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            primes.append(num)
        primeArr.append(num)
    return primes

n = 10000  # example value
print(generate_primes(n))  # output: [2, 3, 5, 7, 11, 13, 17, 19]
print(primes);
```

**prime_diffarance**

```python


for i in range(len(primes)-1):
    diff = primes[i+1] - primes[i]  # calculate the difference
    result.append(diff)  # add the difference to the result array

print(primes); 
print(result)  # output: [2, 2, 2, 2]
```

**Visualization**

```python
xpoints = np.array(primes) #25
ypoints = np.array(result) #24

plt.title("Interval of Prime Numbers")
plt.xlabel("Primes")
plt.ylabel("Defference")
plt.plot(xpoints, ypoints)
plt.show()
```

Max in Result array

```python
def find_max(arr):
    max_num = arr[0]
    for num in arr:
        if num > max_num:
            max_num = num
    return max_num

# arr = [10, 2, 45, 67, 8, 12]  # example array
max_num = find_max(result)
print(max_num)  # output: 67
```
